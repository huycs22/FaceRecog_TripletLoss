```python
def count_missing(series):
    return {
        "null_count": series.isna().sum(),
        "empty_str_count": (series == "").sum() if series.dtype == "object" else 0,
        "empty_list_count": series.apply(lambda x: isinstance(x, list) and len(x) == 0).sum(),
        "empty_dict_count": series.apply(lambda x: isinstance(x, dict) and len(x) == 0).sum(),
    }

def summarize(df, output = None):
    # Apply function to each column
    missing_summary = pd.DataFrame({col: count_missing(df[col]) for col in df.columns}).T
    
    # Add percentage of “missing” values relative to dataset size
    n_rows = len(df)
    missing_summary["%_missing"] = (
        (missing_summary["null_count"] +
         missing_summary["empty_str_count"] +
         missing_summary["empty_list_count"] +
         missing_summary["empty_dict_count"]) / n_rows * 100
    )
    if output:
        missing_summary.to_csv(output)
    print(missing_summary)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModel

device  = torch.device("cuda" if torch.cuda.is_available() else "cpu")
TEXT_MODEL = "bert-base-uncased"
IMG_MODEL = "vit_base_patch16_224"
D_LATENT = 256
BATCH = 128
EPOCHS = 6
TEMPERATURE = 0.07
LEARNING_RATE = 2e-5

class TextEncoder(nn.Module):
    def __init__(self, model_name=TEXT_MODEL, out_dim=256, normalize=True):
        super().__init__()
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.encoder = AutoModel.from_pretrained(model_name)
        self.proj = nn.Linear(self.encoder.config.hidden_size, out_dim)
        self.normalize = normalize

    def forward(self, texts, max_batch_size=1):
        device = next(self.parameters()).device
        all_embs = []
        for i in range(0, len(texts), max_batch_size):
            batch_texts = texts[i:i+max_batch_size]
            batch = self.tokenizer(batch_texts, padding=True, truncation=True, return_tensors='pt').to(device)
            with torch.no_grad():
              outputs = self.encoder(**batch)
            emb = outputs.last_hidden_state.mean(dim=1)  # mean pooling

            emb = self.proj(emb)
            if self.normalize:
                emb = F.normalize(emb, dim=-1)
            all_embs.append(emb)
        return torch.cat(all_embs, dim=0)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import torch

def build_vocab(series, is_list=False):
    """Tạo từ điển {value: index} cho cột categorical"""
    if is_list:
        all_values = set(v for row in series for v in row if v is not None)
    else:
        all_values = set(series.dropna().unique())
    return {val: idx for idx, val in enumerate(sorted(all_values))}

main_cat_vocab = build_vocab(df["main_category"])
store_vocab = build_vocab(df["store"])
categories_vocab = build_vocab(df["categories"], is_list=True)

print("main_cat_vocab:", main_cat_vocab)
print("store_vocab:", store_vocab)
print("categories_vocab size:", len(categories_vocab))
```

```python

def preprocess_tabular(numeric_df, category_df):
    pipeline = make_pipeline(
        SimpleImputer(strategy="median"),
        StandardScaler()
    )
    numeric_processed = pipeline.fit_transform(numeric_df)   # numpy array [n_samples, 2]
    numeric_features = torch.tensor(numeric_processed, dtype=torch.float32)

    # ======== 3. Convert categorical to indices ========
    main_cat_idx = torch.tensor([main_cat_vocab.get(v, 0) for v in category_df["main_category"]], dtype=torch.long)
    store_idx = torch.tensor([store_vocab.get(v, 0) for v in category_df["store"]], dtype=torch.long)
    
    # categories là list nhiều giá trị
    categories_idx_list = [
        torch.tensor([categories_vocab.get(c,0) for c in cats])
        for cats in category_df["categories"]
    ]

    return numeric_features, main_cat_idx, store_idx, categories_idx_list
```

```python
class TabularEncoder(nn.Module):
    def __init__(self,
                 num_numeric_features,
                 main_category_vocab,
                 store_vocab,
                 categories_vocab,
                 embed_dim=32,
                 out_dim=256):
        super().__init__()

        # Embeddings for categorical features
        self.emb_main_cat = nn.Embedding(main_category_vocab, embed_dim)
        self.emb_store = nn.Embedding(store_vocab, embed_dim)
        self.emb_categories = nn.Embedding(categories_vocab, embed_dim)

        # Projection for numeric features
        self.num_proj = nn.Sequential(
            nn.Linear(num_numeric_features, 64),
            nn.ReLU(),
            nn.Linear(64, embed_dim)
        )

        # Fusion layer
        total_dim = embed_dim * 4  # main_cat, store, categories, numeric
        self.fusion = nn.Sequential(
            nn.Linear(total_dim, 128),
            nn.ReLU(),
            nn.Linear(128, out_dim)
        )

    def forward(self, num_features, main_cat_idx, store_idx, categories_idx_list):
        device = next(self.parameters()).device
        num_features = num_features.to(device)
        main_cat_idx = main_cat_idx.to(device)
        store_idx = store_idx.to(device)
        # numeric projection
        n_emb = self.num_proj(num_features)

        # categorical embeddings
        m_emb = self.emb_main_cat(main_cat_idx)
        s_emb = self.emb_store(store_idx)

        # categories may be multiple → average pooling
        cat_embs = []
        for cats in categories_idx_list:
            if len(cats) == 0:
                cat_embs.append(torch.zeros(self.emb_categories.embedding_dim, device=device, requires_grad=True))
            else:
                cats_tensor = torch.tensor(cats, device=device)
                cat_emb = self.emb_categories(cats_tensor)
                cat_embs.append(cat_emb.mean(dim=0))
        cat_embs = torch.stack(cat_embs)

        # concat all
        x = torch.cat([n_emb, m_emb, s_emb, cat_embs], dim=1)

        # fusion
        return F.normalize(self.fusion(x), dim=-1)
```

```python
import requests
import os
import concurrent.futures
from tqdm import tqdm
from urllib.parse import urlparse
from PIL import Image

# --- CẤU HÌNH ---
# Nơi lưu ảnh (trên ổ đĩa Colab)
SAVE_DIR = "/kaggle/working/images"
# Số lượng luồng tải ảnh song song (32 là con số tốt cho Colab)
MAX_WORKERS = 32
# ------------------

def download_image(url_and_path):
    """
    Hàm thực hiện tải 1 ảnh và lưu vào file_path.
    Được thiết kế để chạy trong ThreadPoolExecutor.
    """
    url, file_path = url_and_path

    # Bỏ qua nếu file đã tồn tại (giúp bạn chạy lại script mà không cần tải lại)
    if os.path.exists(file_path):
        return file_path

    try:
        response = requests.get(url, timeout=20)
        # Báo lỗi nếu server trả về 4xx hoặc 5xx
        response.raise_for_status()

        # Ghi file
        with open(file_path, 'wb') as f:
            f.write(response.content)
        return file_path
    except requests.exceptions.RequestException as e:
        print(f"Lỗi khi tải {url}: {e}")
        return None

def preprocess_and_download(main_dataset, save_dir=SAVE_DIR):
    """
    Hàm chính:
    1. Tạo thư mục lưu.
    2. Duyệt qua dataset, tạo danh sách các (url, file_path) cần tải.
    3. Tạo new_main_dataset với 'local_path' mới.
    4. Dùng ThreadPoolExecutor để tải song song.
    """

    print(f"Sẽ lưu ảnh vào thư mục: {save_dir}")
    os.makedirs(save_dir, exist_ok=True)

    tasks = [] # List các (url, file_path) để cho vào thread pool
    new_main_dataset = [] # Dataset mới với local_path

    print("Bước 1/2: Chuẩn bị cấu trúc dữ liệu và danh sách file...")
    # --- Bước 1: Chuẩn bị data ---
    for i, url_dict_list in enumerate(tqdm(main_dataset, desc="Chuẩn bị data")):
        large_urls = [img_dict['large'] for img_dict in url_dict_list if 'large' in img_dict and img_dict['large']]
        new_url_dict_list = []
        for j, img_dict in enumerate(url_dict_list):

            # Giữ nguyên các dict không có 'large'
            new_img_dict = img_dict.copy()

            url = img_dict.get('large')
            if url:
                # Cố gắng giữ phần mở rộng file (ví dụ: .jpg, .png)
                try:
                    path = urlparse(url).path
                    ext = os.path.splitext(path)[1]
                    if not ext: ext = '.jpg' # Mặc định là .jpg
                except:
                    ext = '.jpg'

                file_name = f"item_{i}_img_{j}{ext}"
                file_path = os.path.join(save_dir, file_name)

                # Thêm đường dẫn cục bộ vào dict mới
                new_img_dict['local_path'] = file_path

                # Thêm vào danh sách task
                tasks.append((url, file_path))

            new_url_dict_list.append(new_img_dict)
        new_main_dataset.append(new_url_dict_list)

    print(f"\nChuẩn bị xong. Sẵn sàng tải {len(tasks)} ảnh.")

    # --- Bước 2: Tải ảnh song song ---
    print(f"\nBước 2/2: Bắt đầu tải ảnh với {MAX_WORKERS} luồng...")
    # Dùng list() và tqdm() để hiển thị progress bar
    with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        list(tqdm(executor.map(download_image, tasks), total=len(tasks), desc="Đang tải ảnh"))

    print("\n--- HOÀN TẤT TẢI TRƯỚC ---")
    return new_main_dataset

# --- CÁCH SỬ DỤNG ---

# GIẢ SỬ: `all_my_data` là biến chứa TOÀN BỘ dữ liệu của bạn
# all_my_data = [
#     [{'large': 'http://.../img1.jpg'}, {'large': 'http://.../img2.jpg'}], # item 0
#     [{'large': 'http://.../img3.jpg'}]                                    # item 1
#     # ... (hàng nghìn item khác)
# ]

# Chạy hàm này MỘT LẦN
new_data = preprocess_and_download(df["images"].tolist())

# Kiểm tra kết quả
# print("\nDữ liệu mới:")
# print(new_data[0])
# Sẽ in ra:
# [
#   {'large': 'http://.../img1.jpg', 'local_path': '/content/images/item_0_img_0.jpg'},
#   {'large': 'http://.../img2.jpg', 'local_path': '/content/images/item_0_img_1.jpg'}
# ]
```

```python
    import timm
    class ImageEncoder(nn.Module):
      def __init__(self, model_name=IMG_MODEL, out_dim=256):
        super().__init__()
        self.model = timm.create_model(model_name, pretrained=True)
        # Tắt gradient cho toàn bộ model base
        for param in self.model.parameters():
            param.requires_grad = False
        # Đặt model ở chế độ eval() để tắt Dropout, Batchnorm
        self.model.eval()
        self.proj = nn.Linear(self.model.head.in_features, out_dim)
        self.preprocess = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                std=[0.229, 0.224, 0.225]),
        ])
        self.variant_gate = nn.Sequential(
                nn.Linear(out_dim, out_dim//2),
                nn.ReLU(),
                nn.Linear(out_dim//2, 1)
            )
    
      def forward(self, list_of_url_dicts):
        img_embs = []
        valid_idx = []
        for i, url_dict_list in enumerate(list_of_url_dicts):
          # Trích tất cả URL từ dictionary
          local_paths = [img_dict['local_path'] for img_dict in url_dict_list
                               if 'local_path' in img_dict and img_dict['local_path']]
          if not local_paths or len(local_paths) == 0:
              continue
          else:
              valid_idx.append(i)
              imgs = []
    
              for path in local_paths: # Lặp qua local_paths
                  if not path:
                      continue
                  try:
                      # Gọi hàm read_img đã sửa
                      img_tensor = self.read_img(path)
                      imgs.append(img_tensor)
                  except Exception as e:
                      # Lỗi có thể xảy ra nếu file tải về bị hỏng
                      print(f"Failed to load local file {path}: {e}")
                      continue
    
              if len(imgs) == 0:
                  valid_idx.pop()
                  continue
          batch = torch.stack(imgs).to(device)
          with torch.no_grad():
            feats = self.model.forward_features(batch)
            emb = self.model.forward_head(feats, pre_logits=True)
          out = self.proj(emb)  # [len(imgs), D]
    
          emb_proj = F.normalize(out, dim=-1)
          gates = self.variant_gate(emb_proj)
          weights = F.softmax(gates.squeeze(1), dim=0)
          weighted = (weights.unsqueeze(1) * emb_proj).sum(dim=0)  # [out_dim]
          weighted = F.normalize(weighted, dim=-1)  # chuẩn lại
          img_embs.append(weighted)
    
        B = len(list_of_url_dicts)
        D = self.proj.out_features
    
        if len(img_embs) == 0:
            return torch.zeros((B, D), device=device, dtype=torch.float, requires_grad=True)
    
        # Tạo tensor mới zeros, sau đó scatter từ out vào đúng vị trí
        img_embs_tensor = torch.stack(img_embs).to(device)  # [n_valid, D]
        final = torch.zeros((B, D), device=device, dtype=img_embs_tensor.dtype)
        idx_tensor = torch.tensor(valid_idx, device=device, dtype=torch.long)
        final[idx_tensor] = img_embs_tensor
        return F.normalize(final, dim=-1)
    
    
      def read_img(self, path):
          # Mở file trực tiếp từ đường dẫn
          img = Image.open(path).convert("RGB")
          return self.preprocess(img)
```

```python
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
class MyImageDataset(Dataset):
    def __init__(self, processed_data):
        # processed_data chính là `new_data`
        self.data = processed_data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Trả về một item (chính là cái list url_dict_list)
        return self.data[idx]

def custom_collate(batch):
    """
    Vì mỗi item (list_of_url_dicts) có độ dài khác nhau,
    chúng ta không stack chúng lại.
    Chúng ta chỉ trả về một list các item đó.
    Hàm forward() của bạn đã được thiết kế để xử lý chính xác điều này.
    """
    return batch

train_dataset = MyImageDataset(new_data)
inference_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=False,
    num_workers=4,  # Bắt buộc để tăng tốc
    pin_memory=True,
    collate_fn=custom_collate
)
def get_img_emb(img_encoder, inference_loader):
  all_embeddings = []
  for list_of_url_dicts_batch in tqdm(inference_loader, desc="Đang chạy forward (với grad)"):

      # Hàm forward của bạn sẽ nhận batch này
      # Đồ thị tính toán SẼ được tạo ra cho proj và gate
      final_embs = img_encoder(list_of_url_dicts_batch)

      # KHÔNG DÙNG .cpu()
      # Giữ tensor trên GPU với đồ thị tính toán của nó
      all_embeddings.append(final_embs)

  # Nối tất cả kết quả lại
  # final_tensor này vẫn còn "sống" và có thể gọi .backward()
  img_emb = torch.cat(all_embeddings, dim=0)
  return img_emb

img_encoder = ImageEncoder().to(device)
img_encoder.train()
img_emb = get_img_emb(img_encoder, inference_loader)
```

```python
class ContrastiveLearning(nn.Module):
  def __init__(self, temperature):
      super().__init__()
      self.temperature = temperature

  def forward(self, v1, v2):
    v1 = v1.to(device)
    v2 = v2.to(device)

    sim_v1_v2 = torch.matmul(v1, v2.T) / self.temperature
    labels = torch.arange(sim_v1_v2.shape[0], device=device)
    loss_v1_2_v2 = F.cross_entropy(sim_v1_v2, labels)
    loss_v2_2_v1 = F.cross_entropy(sim_v1_v2.T, labels)

    loss = (loss_v1_2_v2 + loss_v2_2_v1) / 2
    return loss
```

```python
from torch.utils.data import Dataset
import torch

class CombinedDataset(Dataset):
    def __init__(self, img_data, text_data, numeric_data, category_data):
        self.img_data = img_data
        self.text_data = text_data
        self.numeric_data = numeric_data
        self.category_data = category_data
        assert len(self.img_data) == len(self.text_data) == len(self.numeric_data)

    def __len__(self):
        # Trả về tổng số item
        return len(self.text_data)

    def __getitem__(self, idx):
        # Lấy dữ liệu cho item thứ 'idx'
        
        img_item = self.img_data[idx]
        text_item = self.text_data[idx]
        
        # Lấy dữ liệu tabular
        num_item = self.numeric_data.iloc[idx]
        # Lấy item thứ idx từ tensor categorical
        cat_items = self.category_data.iloc[idx]

        return (img_item, text_item, num_item, cat_items)

def training_collate_fn(batch):
    img_items_list = []
    text_items_list = []
    
    # Tách dữ liệu tabular
    numeric_batch = []
    cat_batch = []
    
    for (img_item, text_item, num_item, cat_items) in batch:
        img_items_list.append(img_item)
        text_items_list.append(text_item)
        numeric_batch.append(num_item)
        cat_batch.append(cat_items)

    return img_items_list, text_items_list,num_item, cat_items

train_dataset = CombinedDataset(
    new_data, 
    df["text_combined"].tolist(),
    numeric_df,
    category_df
)

BATCH_SIZE = 2
train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True, # Bật xáo trộn cho training
    num_workers=4,
    pin_memory=True,
    collate_fn=training_collate_fn # Sử dụng hàm gộp tùy chỉnh
)
```

```python
# --- 1. MODIFIED TextEncoder ---
# (Replace the class in Cell 55 with this)

class TextEncoder(nn.Module):
    def __init__(self, model_name=TEXT_MODEL, out_dim=D_LATENT, normalize=True):
        super().__init__()
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.encoder = AutoModel.from_pretrained(model_name)
        self.proj = nn.Linear(self.encoder.config.hidden_size, out_dim)
        self.normalize = normalize

    # --- MODIFIED FORWARD METHOD ---
    def forward(self, batch_texts):
        """
        Processes a mini-batch of texts (list of strings)
        """
        device = next(self.parameters()).device
        
        # Tokenize the whole batch at once
        batch = self.tokenizer(batch_texts, padding=True, truncation=True, return_tensors='pt').to(device)
        
        # Run base encoder with no_grad
        with torch.no_grad():
          outputs = self.encoder(**batch)
        
        # Mean pooling
        emb = outputs.last_hidden_state.mean(dim=1)

        # Pass through projection head (this part IS trained)
        emb = self.proj(emb)
        if self.normalize:
            emb = F.normalize(emb, dim=-1)
        return emb

# --- 2. MODIFIED TabularEncoder ---
# (Replace the class in Cell 16 with this)

class TabularEncoder(nn.Module):
    def __init__(self,
                 num_numeric_features,
                 main_category_vocab,
                 store_vocab,
                 categories_vocab,
                 embed_dim=32,
                 out_dim=D_LATENT): # Use constant
        super().__init__()

        # Embeddings for categorical features
        self.emb_main_cat = nn.Embedding(main_category_vocab, embed_dim)
        self.emb_store = nn.Embedding(store_vocab, embed_dim)
        self.emb_categories = nn.Embedding(categories_vocab, embed_dim)

        # Projection for numeric features
        self.num_proj = nn.Sequential(
            nn.Linear(num_numeric_features, 64),
            nn.ReLU(),
            nn.Linear(64, embed_dim)
        )

        # Fusion layer
        total_dim = embed_dim * 4  # main_cat, store, categories, numeric
        self.fusion = nn.Sequential(
            nn.Linear(total_dim, 128),
            nn.ReLU(),
            nn.Linear(128, out_dim)
        )

    # --- MODIFIED FORWARD METHOD ---
    def forward(self, num_features, main_cat_idx, store_idx, categories_idx_list):
        device = next(self.parameters()).device
        
        # Move tensors to device
        num_features = num_features.to(device)
        main_cat_idx = main_cat_idx.to(device)
        store_idx = store_idx.to(device)
        
        # numeric projection
        n_emb = self.num_proj(num_features)

        # categorical embeddings
        m_emb = self.emb_main_cat(main_cat_idx)
        s_emb = self.emb_store(store_idx)

        # categories may be multiple → average pooling
        cat_embs = []
        for cats in categories_idx_list: # 'cats' is a tensor from the batch
            # Move the tensor to the device
            cats = cats.to(device)
            
            if len(cats) == 0:
                # Use a zero tensor if no categories
                cat_embs.append(torch.zeros(self.emb_categories.embedding_dim, device=device))
            else:
                # 'cats' is already the tensor of indices
                cat_emb = self.emb_categories(cats)
                cat_embs.append(cat_emb.mean(dim=0))
        
        # Stack the mean embeddings for the batch
        cat_embs = torch.stack(cat_embs)

        # concat all
        x = torch.cat([n_emb, m_emb, s_emb, cat_embs], dim=1)

        # fusion
        return F.normalize(self.fusion(x), dim=-1)
```

```python
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
from tqdm.auto import tqdm

# --- 1. The Dataset Class ---
class MultiModalDataset(Dataset):
    def __init__(self, texts, images_data, numeric_features, main_cat_idx, store_idx, categories_idx_list):
        # Ensure all data is of the same length
        assert len(texts) == len(images_data) == len(numeric_features) == len(main_cat_idx)
        
        self.texts = texts # List of strings
        self.images_data = images_data # List of (list of dicts)
        self.numeric_features = numeric_features # Tensor [N, num_features]
        self.main_cat_idx = main_cat_idx # Tensor [N]
        self.store_idx = store_idx # Tensor [N]
        self.categories_idx_list = categories_idx_list # List of Tensors
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        # Return all data for one item
        return (
            self.texts[idx],
            self.images_data[idx],
            self.numeric_features[idx],
            self.main_cat_idx[idx],
            self.store_idx[idx],
            self.categories_idx_list[idx]
        )

# --- 2. The Custom Collate Function ---
def custom_collate(batch):
    """
    Takes a list of items from __getitem__ and bundles them into a batch.
    """
    # Unzip the batch
    texts = [item[0] for item in batch]
    images_data = [item[1] for item in batch]
    numeric_features = torch.stack([item[2] for item in batch])
    main_cat_idx = torch.stack([item[3] for item in batch])
    store_idx = torch.stack([item[4] for item in batch])
    categories_idx_list = [item[5] for item in batch] # This is a list of lists/tensors
    
    return texts, images_data, numeric_features, main_cat_idx, store_idx, categories_idx_list

# --- 3. Instantiate Dataset and DataLoader ---

# Get the list of texts from the DataFrame
texts_list = df["text_combined"].tolist()

# (Assuming 'new_data', 'numeric_features', 'main_cat_idx', 'store_idx', 
# and 'categories_idx_list' are all available from your notebook)

full_dataset = MultiModalDataset(
    texts_list, 
    new_data, 
    numeric_features, 
    main_cat_idx, 
    store_idx, 
    categories_idx_list
)

train_loader = DataLoader(
    full_dataset,
    batch_size=BATCH, # Your constant (128)
    shuffle=True,
    collate_fn=custom_collate,
    num_workers=2, # Use 2 workers for loading
    pin_memory=True
)

print(f"DataLoader created with {len(train_loader)} batches.")
```

```python
# --- 1. Initialize Models ---
# (Make sure the MODIFIED TextEncoder and TabularEncoder are defined)

text_encoder = TextEncoder().to(device)

img_encoder = ImageEncoder().to(device) # From Cell 49

tab_encoder = TabularEncoder(
     num_numeric_features=numeric_features.shape[1], # Use shape for safety
     main_category_vocab=len(main_cat_vocab),
     store_vocab=len(store_vocab),
     categories_vocab=len(categories_vocab),
     embed_dim=32,
     out_dim=D_LATENT
).to(device)

# --- 2. Optimizer and Criterion ---
# (Collect parameters from ALL 3 models)
optimizer = optim.Adam(
    list(text_encoder.parameters()) +
    list(img_encoder.parameters()) +
    list(tab_encoder.parameters()),
    lr=LEARNING_RATE # Your constant
)

criterion = ContrastiveLearning(temperature=TEMPERATURE) # Your constant

# --- 4. The Training Loop ---
print("--- Starting Training ---")

for epoch in range(EPOCHS):
    
    # Set models to train mode
    text_encoder.train()
    img_encoder.train()
    tab_encoder.train()
    
    # Use tqdm for a progress bar
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}")
    total_loss = 0.0
    
    for batch in loop:
        # Unpack the mini-batch
        texts_b, images_b, numeric_b, main_cat_b, store_b, categories_b = batch
        
        # --- Forward Pass ---
        # Get embeddings for the current mini-batch
        text_emb = text_encoder(texts_b)
        img_emb = img_encoder(images_b)
        tab_emb = tab_encoder(numeric_b, main_cat_b, store_b, categories_b)
        
        # --- Calculate Loss ---
        # (This is a 3-way contrastive loss)
        loss_1 = criterion(img_emb, text_emb)
        loss_2 = criterion(img_emb, tab_emb)
        loss_3 = criterion(text_emb, tab_emb)
        
        # Sum the losses
        loss = loss_1 + loss_2 + loss_3
        
        # --- Backward Pass & Optimize ---
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        loop.set_postfix(loss=loss.item()) # Show current batch loss
    
    avg_loss = total_loss / len(train_loader)
    print(f"End of Epoch {epoch+1}: Avg. Loss = {avg_loss:.4f}")
            
print("--- Training Finished ---")
```

```python
class FusionGate(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.gate_mlp = nn.Sequential(
            nn.Linear(dim, dim//2),
            nn.ReLU(),
            nn.Linear(dim//2, 1)
        )

    def forward(self, e_text, e_img, e_tab, mask=None):
        # e_text, e_img, e_tab: [B, D]
        B, D = e_text.size()
        # tính score gating
        g_t = self.gate_mlp(e_text)  # [B,1]
        g_i = self.gate_mlp(e_img)
        g_b = self.gate_mlp(e_tab)
        gates = torch.cat([g_t, g_i, g_b], dim=1)  # [B,3]

        # nếu có mask, làm cho những modality thiếu có score = -inf trước softmax
        if mask is not None:
            # mask type bool: True = modality có, False = thiếu
            neg_inf = torch.tensor(-1e9, device=gates.device, dtype=gates.dtype)
            # gates = gates.masked_fill(~mask, neg_inf)
            gates = torch.where(mask, gates, neg_inf)

        # softmax để có trọng số mỗi modal (tổng =1)
        weights = F.softmax(gates, dim=1)  # [B,3]

        # weighted sum các embedding
        fused = (weights[:,0:1] * e_text
                 + weights[:,1:2] * e_img
                 + weights[:,2:3] * e_tab)  # [B, D]

        # chuẩn hóa nếu cần
        fused = F.normalize(fused, dim=-1)

        return fused.to(device), weights
```

```python
img_missing_mask = (img_embeddings_all.abs().sum(dim=1) == 0)
text_missing_mask = (text_embeddings_all.abs().sum(dim=1) == 0)
tab_missing_mask = (tab_embeddings_all.abs().sum(dim=1) == 0)

B, D = img_embeddings_all.shape

modal_mask = torch.ones((B,3), dtype=torch.bool, device=device)
modal_mask[:,0] = ~text_missing_mask
modal_mask[:,1] = ~img_missing_mask
modal_mask[:,2] = ~tab_missing_mask

fusion_gate = FusionGate(D).to(device)
fused_emb, weights = fusion_gate(text_embeddings_all, img_embeddings_all, tab_embeddings_all, mask=modal_mask)
```
